```python
df = pd.read_csv("/kaggle/input/ipl-matchwise-scorecard-2008-2023/Dataset/batting_summary.csv")
df
```

**Space Fixing**

```python
# Space available in columns lets fix it first:
df.columns = df.columns.str.replace(" ","_")
df.columns
```

**Total Match Played**

```python
#each row potray data for each match. 
#total row number is the total match number

num_rows = len(df)
print('Total Match Played:',num_rows)
```

**Total Team Played**

```python
num_unique_col1 = df['Team_Batting'].nunique()
print('Total Team Played:',num_unique_col1)
```

**Max Total Run of a Team**

```python
# Find the row with the maximum score
max_score_row = df[df['Total_Runs'] == df['Total_Runs'].max()]

# Extract person name and score
Team_Batting = max_score_row['Team_Batting'].values[0]
max_score = max_score_row['Total_Runs'].values[0]

# Print the result
print(f"Team with the max Runs: {Team_Batting} (Score: {max_score})")
```

**Total Run of Each Team**

```python
# Pivot table to get total score per team
Run_per_Team = df.pivot_table(values='Total_Runs', index='Team_Batting', aggfunc='sum')

# Print the pivot table
print(Run_per_Team)

# Store the pivot table in a variable (create a copy to avoid modifying the original)
stored_team_pivot_table = Run_per_Team.copy()
```

**Each Batsman Total Run**

```python
# Pivot table to get total score per batsman
Run_per_Batsman = df.pivot_table(values='Runs_Scored', index='Batsman_Names', aggfunc='sum')

# Print the pivot table
print(Run_per_Batsman)

# Store the pivot table in a variable (create a copy to avoid modifying the original)
stored_pivot_table = Run_per_Batsman.copy()
```

**Max Run among Batsman**

```python
# 3. Calculate maximum value and its index (batsman name)
max_value = stored_pivot_table.max()
max_scorer_name = max_value.index[0]
print(f"Max Scorer: {max_scorer_name} with {max_value} runs")
```

**Average Runs Per Batsman**

```python
df['Runs_Scored'] = pd.to_numeric(df['Runs_Scored'], errors='coerce')  # Replace 'NA' with NaN
df.fillna(df['Runs_Scored'].mean(), inplace=True)  # Impute NaN with average score
Average_Runs_per_Batsman = df.pivot_table(values='Runs_Scored', index='Batsman_Names', aggfunc='mean')
print(Average_Runs_per_Batsman)
```
